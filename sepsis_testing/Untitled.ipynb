```python
from sklearn.tree import DecisionTreeRegressor 
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/agesak/Documents/MPH/Thesis/decision_tree/all_locations_redistributed.csv")
```

```python
# view all the elements in the df columns
pd.set_option('display.max_columns', 500)
```

```python
df.head()
```

```python
df["sepsis_cases"] = df.cases_explicit.fillna(0) + df.cases_implicit.fillna(0)
df["sepsis_deaths"] = df.deaths_explicit.fillna(0) + df.deaths_implicit.fillna(0)
df["sepsis_cfr"] = df.sepsis_deaths/df.sepsis_cases
```

```python
df = df.groupby(['sex_id', 'age_group_id', 'haqi', 'location_id', 'year_id', 'level_1', 'level_2'])["sepsis_cfr"].sum().reset_index()
```

```python
df.head()
```

```python
[x for x in list(df) if ("sepsis" not in x) & ("level_1" not in x)]
```

```python
X = df[["age_group_id", "sex_id", "haqi", "level_2"]]
Y = df['sepsis_cfr']
```

```python
dtree=DecisionTreeRegressor()
dtree.fit(X, Y)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
